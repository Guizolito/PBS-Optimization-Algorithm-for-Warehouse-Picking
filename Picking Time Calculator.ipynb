```python
#Atribuição de posições aos diversos SKUs

import pandas as pd
import math
AeB = pd.read_csv('2.1SKUs.csv')
BUs = pd.read_csv('BU.csv')
AeB['x'] = ""
AeB['y'] = ""

xx = []
yy = []
NbrOfItems = 121
for ind,bu in BUs.iterrows():
    BUconsiderado = bu['BU']
    SKUSCons = AeB.loc[AeB['BU'] == BUconsiderado] 
    for index,row in SKUSCons.iterrows():
            
        RowInWarehouse = index / NbrOfItems

        xx.append(math.floor(RowInWarehouse))
        yy.append(round((RowInWarehouse - math.floor(RowInWarehouse)) * NbrOfItems))
AeB['x'] = xx
AeB['y'] = yy

AeB.to_csv('2.1OSKUs.csv')
```
